```python
def plot(metric_name, selected_df, x, probability=False, x_label=None, y_label=None):
    for local_planner_node in selected_df.local_planner_node.unique():
        plot_df = selected_df[selected_df.local_planner_node == local_planner_node].groupby(by=x)[metric_name].agg('mean')

        plt.plot(plot_df.index, plot_df, local_planner_symbol[local_planner_node] + '-', label=local_planner_node.upper(), color=local_planner_color[local_planner_node])
    
    x_values = selected_df[x].unique()
    x_values = [item for item in x_values if not(math.isnan(item)) == True] # remove nan values
#     if 0 not in x_values:  # if 0 is not in the list of values, add it so that every x-axis starts from 0
#         x_values = np.append(x_values, 0)
    
    plt.xticks(x_values)
    
    if probability:
        plt.yticks(np.arange(0, 1.1, 0.1))
    
    plt.legend(fontsize=14)
    plt.grid()
    
    plt.xlabel(x_label, fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    
    out_dir = "MetroXRAINE"
    if not path.exists(out_dir):
        os.makedirs(out_dir)
    plt.savefig(f"{out_dir}/success-rate---num-pedestrians.pdf", bbox_inches='tight')
    plt.show()


df_ped = df[(df.min_turning_radius != 0.0) | (df.local_planner_node != "teb")].copy() # discard teb runs with min turning radius = 0
df_ped_group = df_ped[((df_ped.include_costmap_obstacles == True) & (df_ped.include_dynamic_obstacles == True)) | (df_ped.local_planner_node != "teb")]

plot(metric_name='success_rate', selected_df=df_ped_group, x='pedestrian_number', probability=True, x_label="number of pedestrians", y_label="local planner success rate")
```
