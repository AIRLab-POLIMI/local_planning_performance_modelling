```python
def plot(metric_name, selected_df, x, probability=False):
    for local_planner_node in selected_df.local_planner_node.unique():
        plot_df = selected_df[selected_df.local_planner_node == local_planner_node].groupby(by=x)[metric_name].agg('mean')

        plt.plot(plot_df.index, plot_df, local_planner_symbol[local_planner_node] + '-', label=local_planner_node, color=local_planner_color[local_planner_node])
    
    x_values = selected_df[x].unique()
    x_values = [item for item in x_values if not(math.isnan(item)) == True] # remove nan values
#     if 0 not in x_values:  # if 0 is not in the list of values, add it so that every x-axis starts from 0
#         x_values = np.append(x_values, 0)
    
    plt.xticks(x_values)
    
    if probability:
        plt.yticks(np.arange(0, 1.1, 0.1))
    
    plt.legend()
    
    x_label = x.replace('_', ' ').replace(',', ', ').title()
    y_label = metric_name.replace('_', ' ').replace(',', ', ').title()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    title = x_label + ' vs ' + y_label
    plt.title(title)
    plt.grid()
    plt.savefig('images/' + metric_name + '_' + x + '.png')
    plt.show()
    
    
    
```

```python
# pedestrian_number
plot(metric_name='success_rate', selected_df=df, x='pedestrian_number', probability=True)
plot(metric_name='collision_rate', selected_df=df, x='pedestrian_number', probability=True)
plot(metric_name='execution_time', selected_df=df, x='pedestrian_number')
plot(metric_name='trajectory_length', selected_df=df, x='pedestrian_number')
plot(metric_name='minimum_clearance', selected_df=df, x='pedestrian_number')
plot(metric_name='normalized_curvature', selected_df=df, x='pedestrian_number')
```

```python
def bar_plot(metric_name, selected_df, y_columns=['local_planner_node'], percentage=False, unit_symbol=None, higher_better=True):
    bar_df = selected_df.groupby(y_columns)[metric_name].agg(['mean', 'count']).sort_values(by=('mean'), ascending=higher_better)
    ax = (bar_df.sort_values(by=('mean'), ascending=higher_better)* (100 if percentage else 1) ).plot.barh(y=[('mean')], stacked=True, figsize=[10, 5], grid=True, legend=False)
    ax.set_xlabel(metric_name.replace('_', ' ').replace(',', ', ') + (f" [{unit_symbol}]" if unit_symbol else ""))
    ax.set_ylabel(ax.get_ylabel().replace('_', ' ').replace(',', ',\n'))
    plt.gcf().set_size_inches([5, 1])
    plt.savefig('images/' + metric_name + '_bar_plot.png')
    plt.show()
    
printmd("## success rate")
bar_plot(metric_name='success_rate', selected_df=df, percentage=True, higher_better=True, unit_symbol='%')
printmd("## collision rate")
bar_plot(metric_name='collision_rate', selected_df=df, percentage=True, higher_better=True, unit_symbol='%')
printmd("## execution time")
bar_plot(metric_name='execution_time', selected_df=df, higher_better=False, unit_symbol='s')
printmd("## trajectory length")
bar_plot(metric_name='trajectory_length', selected_df=df, higher_better=False, unit_symbol='m/s')
printmd("## minimum clearance")
bar_plot(metric_name='minimum_clearance', selected_df=df, higher_better=False, unit_symbol='m')
printmd("## normalized curvature")
bar_plot(metric_name='normalized_curvature', selected_df=df, higher_better=False)
printmd("## movebase cpu time")
bar_plot(metric_name='move_base_cpu_time', selected_df=df, higher_better=True, unit_symbol='s')
printmd("## movebase max memory")
bar_plot(metric_name='move_base_max_memory', selected_df=df, higher_better=True, unit_symbol='Mb')
```
